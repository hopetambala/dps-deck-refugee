```python
import numpy as np
import pandas as pd
import string

files = {'asia_pacific': '2_RSQSubmissions_asia_pacific.csv',
         'eastafrica': '2_RSQSubmissions_eastafrica.csv',
         'europe': '2_RSQSubmissions_europe.csv',
         'middleeast_northafrica': '2_RSQSubmissions_middleeast_northafrica.csv',
         'southernafrica': '2_RSQSubmissions_southernafrica.csv',
         'westcentralafrica': '2_RSQSubmissions_westcentralafrica.csv',
         'countries_codes_and_coordinates.csv': 'countries_codes_and_coordinates.csv' }
df = pd.read_csv(files['asia_pacific'])
df2 = pd.read_csv(files['eastafrica'])
df3 = pd.read_csv(files['europe'])
df4 = pd.read_csv(files['middleeast_northafrica'])
df5 = pd.read_csv(files['southernafrica'])
df6 = pd.read_csv(files['westcentralafrica'])

df_country = pd.read_csv(files['countries_codes_and_coordinates.csv'])
```

```python
columns = ['Year', 'Country of Origin (ISO)', 'Country of Origin',
       'Latitude - Country of Origin', 'Longitude - Country of Origin',
       'Country of Asylum (ISO)', 'Country of Asylum',
       'Latitude - Country of Asylum', 'Longitude - Country of Asylum',
       'Country of Resettlement (ISO)', 'Country of Resettlement',
       'Latitude - Country of Resettlement',
       'Longitude - Country of Resettlement', 'Total departures (persons)']
```

```python
for col in df_country.columns: 
    #if df_country[col].dtype != int:
    df_country[col] = df_country[col].astype(str).str.replace("\"",'')

# for col in df3.columns: 
#     try:
#         df3[col] = df3[col].str.replace('[^\w\s]','')
#     except:
#         pass
    
df_country.head()
```

```python
def geocode(country_name):
    for index, row in df_country.iterrows():
        if country_name == row['Alpha-3code']: 
            arr = np.array([float(row['Longitude (average)']),float(row['Latitude (average)'])])      
            lisst = list(arr)
            return lisst
            
            
```

```python
list_of_datasets = [df, df2, df3,df4,df5,df6]
```

```python
for index, dataset in enumerate(list_of_datasets):
    dataset['Country of Origin Coordinates'] = dataset['Country of Origin (ISO)'].apply(geocode)
    dataset['Country of Asylum Coordinates'] = dataset['Country of Asylum (ISO)'].apply(geocode)
    dataset['Country of Resettlement Coordinates'] = dataset['Country of Resettlement (ISO)'].apply(geocode)  
    
```

```python
for index, dataset in enumerate(list_of_datasets):
    dataset['Country of Origin Coordinates'].fillna(value=pd.Series(np.array([0,0])), inplace=True)
    dataset['Country of Asylum Coordinates'].fillna(value=pd.Series(np.array([0,0])), inplace=True)
    dataset['Country of Resettlement Coordinates'].fillna(value=pd.Series(np.array([0,0])), inplace=True)
```

```python
list_of_datasets
```

```python
for index, dataset in enumerate(list_of_datasets):
    dataset.to_csv("cleaned/2_RSQSubmissions_"+str(index)+"_cleaned.csv")
    dataset.to_json("cleaned/2_RSQSubmissions_"+str(index)+"_cleaned.json", orient='records')
```

```python
megadf = pd.concat(list_of_datasets, ignore_index=True)
megadf.to_csv("cleaned/2_RSQSubmissions_MEGA_cleaned.csv")
megadf.to_json("cleaned/2_RSQSubmissions_MEGA_cleaned.json", orient='records')
```
