```python
import numpy as np
import pandas as pd
import string

files = {'1_RSQDepartures.csv': '1_RSQDepartures.csv','2_RSQSubmissions.csv': '2_RSQSubmissions.csv','3_RSQDemographics.csv': '3_RSQDemographics.csv','countries_codes_and_coordinates.csv': 'countries_codes_and_coordinates.csv' }
df = pd.read_csv(files['1_RSQDepartures.csv'])
df2 = pd.read_csv(files['2_RSQSubmissions.csv'])
df3 = pd.read_csv(files['3_RSQDemographics.csv'])

df_country = pd.read_csv(files['countries_codes_and_coordinates.csv'])
```

```python
for col in df_country.columns: 
    #if df_country[col].dtype != int:
    df_country[col] = df_country[col].astype(str).str.replace("\"",'')

# for col in df3.columns: 
#     try:
#         df3[col] = df3[col].str.replace('[^\w\s]','')
#     except:
#         pass
    
df_country.head()
```

```python
def geocode(country_name):
    for index, row in df_country.iterrows():
        if country_name == row['Country']: 
            return np.array([float(row['Longitude (average)']),float(row['Latitude (average)'])])
```

```python
df2['Country of Origin Coordinates'] = df2['Country of Origin'].apply(geocode)
df2['Country of Asylum Coordinates'] = df2['Country of Asylum'].apply(geocode)
df2['Country of Resettlement Coordinates'] = df2['Country of Resettlement'].apply(geocode)

df3['Country of Origin Coordinates'] = df3['Country of Origin'].apply(geocode)
df3['Country of Resettlement Coordinates'] = df3['Country of Resettlement'].apply(geocode)

# df['Country of Origin Coordinates'] = df['Country of Origin (ISO)'].apply(geocodeISO)
# df['Country of Asylum Coordinates'] = df['Country of Asylum'].apply(geocodeISO)
# df['Country of Resettlement Coordinates'] = df['Country of Resettlement (ISO)'].apply(geocodeISO)
```
